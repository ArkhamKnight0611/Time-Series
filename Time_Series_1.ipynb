Q1. Time Series and Applications:

A time series is a sequence of data points collected at regular intervals over time.
Common applications of time series analysis include:
Forecasting: Predicting future sales, stock prices, or weather patterns.
Trend analysis: Identifying long-term changes in data.
Seasonality detection: Uncovering recurring patterns within a year, month, or week.
Anomaly detection: Spotting unusual deviations from the expected pattern.
Q2. Common Time Series Patterns:

Trend: A general upward or downward movement over time.

Seasonality: Repeating patterns within a specific time period (e.g., daily, weekly, yearly).

Cyclicity: Long-term fluctuations that rise and fall over extended periods.

Stationarity: Consistent statistical properties (mean, variance) over time.

These patterns can be identified through visual inspection of time series plots and statistical techniques like autocorrelation analysis.

Interpretation: Trends suggest future values will continue in that direction. Seasonality helps predict future values based on the seasonal cycle. Cyclicity helps anticipate future peaks and troughs.

Q3. Time Series Preprocessing:

Data cleaning: Handling missing values and outliers.
Transformation: Scaling or normalization to ensure consistent data units.
Differencing: Removing trends or seasonality by subtracting previous values.
Stationarity checks: Ensuring the data's statistical properties are stable over time.
Q4. Time Series Forecasting in Business:

Businesses use forecasting to:

Plan inventory and resources.
Develop marketing campaigns.
Manage cash flow.
Make strategic decisions based on predicted future demand.
Challenges:

Data accuracy and completeness.
Unexpected events impacting future trends.
Difficulty in forecasting long-term horizons.
Q5. ARIMA Modeling:

ARIMA (Autoregressive Integrated Moving Average) is a popular time series forecasting model.
It combines:
Autoregression (AR): Uses past values to predict future values.
Integration (I): Makes the data stationary through differencing.
Moving Average (MA): Considers the average of past forecast errors.
Q6. ACF and PACF for ARIMA Order Identification:

Autocorrelation Function (ACF) plot shows the correlation between a time series and its lagged versions.
Partial Autocorrelation Function (PACF) helps identify the number of past values (AR terms) needed in the model.
By analyzing the decay patterns in ACF and PACF plots, we can determine the appropriate order for the ARIMA model (ARIMA(p, d, q)).
Q7. ARIMA Model Assumptions:

Stationarity: The data's statistical properties are constant over time.

Normality: The residuals (errors) follow a normal distribution.

No autocorrelation in residuals: Past forecast errors are not correlated with current errors.

These assumptions can be tested using statistical tests like Dickey-Fuller test for stationarity and Shapiro-Wilk test for normality.

Q8. Forecasting Retail Sales:

For monthly sales data, a Seasonal ARIMA model (SARIMA) would be recommended.
SARIMA accounts for both trend and seasonal patterns present in monthly data.
Q9. Limitations of Time Series Analysis:

Limited ability to predict unexpected events or structural changes.

Relies heavily on historical data, assuming future patterns resemble the past.

Model accuracy deteriorates as forecasting horizons become longer.

Example: A sudden change in consumer preferences due to a new competitor might not be reflected in historical sales data, leading to inaccurate forecasts.

Q10. Stationarity and Forecasting Models:

A stationary time series has constant mean, variance, and covariance over time.
Most forecasting models, including ARIMA, require stationary data.
Non-stationary data can lead to spurious (false) relationships and unreliable forecasts.
Differencing is often used to transform non-stationary data into a stationary form.
